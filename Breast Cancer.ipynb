## BREAST CANCER CLASSIFICATION

# STEP #1: PROBLEM STATEMENT

- Predicting if the cancer diagnosis is benign or malignant based on several observations/features 
- 30 features are used, examples:
        - radius (mean of distances from center to points on the perimeter)
        - texture (standard deviation of gray-scale values)
        - perimeter
        - area
        - smoothness (local variation in radius lengths)
        - compactness (perimeter^2 / area - 1.0)
        - concavity (severity of concave portions of the contour)
        - concave points (number of concave portions of the contour)
        - symmetry 
        - fractal dimension ("coastline approximation" - 1)

- Datasets are linearly separable using all 30 input features
- Number of Instances: 569
- Class Distribution: 212 Malignant, 357 Benign
- Target class:
         - Malignant
         - Benign

# STEP #2: IMPORTING DATA

```python
#import libraries data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#allow to show matplotlib plots in notebook
%matplotlib inline
```

```python
#import data from scikit-learn library datasets
from sklearn.datasets import load_breast_cancer
```

```python
#create object of load_breast_cancer to fetch data
cancer_data = load_breast_cancer()
```

```python
#check values in cancer_data object
cancer_data
```

```python
#store object values in pandas dataframe for easier manipulation
df = pd.DataFrame(data= np.c_[cancer_data['data'], cancer_data['target']],
                     columns= cancer_data['feature_names'].tolist() + ['target'])
```

```python
df.head()
```

```python
df.tail()
```

```python
#Currently target values are in float
#making all values in target column as integer
df['target'] = df['target'].apply(int)
```

# STEP #3: VISUALIZING THE DATA

```python
#Lets visualise data b/w multiple features
#0 is 'malignant' and 1 is 'benign' which can be found from 'target_names' in cancer_data above
sns.pairplot(data = df, vars= ['mean radius','mean texture','mean perimeter','mean smoothness','mean compactness'], hue = 'target')
```

```python
#in above plot we can understand relation b/w features 
#if you look into mean radius vs mean perimeter
sns.scatterplot(x='mean radius',y='mean perimeter',data = df,hue='target')
#it seams when mean radius > 15 and mean perimeter > 100 then cancer if generally found 'malignant' otherwise 'benign' 
```

```python
#lets check how many of our data is malignant and how many 'benign'
sns.countplot(df['target'])
#we have around 210 malignant data and around 350 benign data
```

# STEP #4: MODEL TRAINING

our features are varying with different ranges. Lets try without normalization of training our Machine Learning Model. We will be using Support Vector Machine Algorithm for classification. 

## Split data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target',axis=1)
```

```python
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)
```

## Train Model in SVM

```python
#importing SVM algorithm class
from sklearn.svm import SVC
```

```python
#creating object of SVC
#using default C, gamma values. by default C = 1.0
svm = SVC(gamma = 'auto')
```

```python
#Lets fit our train data to svm algorithm
svm.fit(X_train,y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
#importing classes to evaluate our above model
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#predicting values of target of test data
predictions = svm.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test,predictions),annot = True)
#True Positive(benign) are all correct but True Negative(Malign) are all wrong
#For all inputs our model is giving 1 value
```

```python
#classification Matrix to see our accuracy of model
print(classification_report(y_test,predictions))
#precision is only 41% which is very low
```

# STEP #6: IMPROVING THE MODEL

Lets try to improve our Model
Issues we can see
1. Feature Scaling - our columns have different range of values. lets try to bring it down to small values in all features
2. Better C and gamma values to improve Optimization and Regularization

```python
#Normalizing our data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled_df = pd.DataFrame(data= X_scaled, columns= X.columns)
```

```python
X_scaled_df.head()
```

```python
#Lets split scaled data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.6, random_state=42)
```

```python
#Lets apply our scaled data to fit our SVM algorithm
svm_scaled = SVC(gamma= 'auto')
```

```python
svm_scaled.fit(X_train,y_train)
```

```python
#Predicting our new model with scaled input data
predictions = svm_scaled.predict(X_test)
```

```python
#Evaluating our new model
print(confusion_matrix(y_test,predictions))
#True Negative(Malignant) are 119 correct and only 4 wrong
#True Positive(Benign) are 213 correct and only 6 wrong
```

```python
print(classification_report(y_test,predictions))
#we have 97% precision. Very good algorithm.
```

```python
# Lets see if we can improve this model using best fit for C and gamma
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C": [0.1,0.3,1,3,10,30,100],
    "gamma": [1,0.3,0.1,0.03,0.01],
    "kernel": ['rbf']
}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,verbose=3,cv = 5)
grid.fit(X_train,y_train)
```

```python
#checking best parameters
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```

With Scaled Features and Best C, gamma values we have achieved a 98% precision
